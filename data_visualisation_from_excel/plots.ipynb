```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Financial Sample.csv")
```

```python
df.head()
```

```python
df_vtt_canada = df.loc[(df['Country']== 'Canada') & (df['Product']=='VTT') & (df['Segment']=='Government')]
df_vtt_canada = df_vtt_canada.sort_values(by=['Date'])

df_vtt_france = df.loc[(df['Country']== 'France') & (df['Product']=='VTT') & (df['Segment']=='Government')]
df_vtt_france = df_vtt_france.sort_values(by=['Date'])

df_vtt_canada.plot(x='Date', y='Profit', rot=90)
plt.title('Canadian profits for VTT over time')
df_vtt_france.plot(x='Date', y='Profit', rot=45)
plt.title('Canadian profits for VTT over time')
plt.title('French profits for VTT over time')
```

```python
df_products = df.groupby(['Product']).sum()
df_products['Units Sold'].plot.bar(rot=45)
```
