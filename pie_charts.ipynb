```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
slices = [60, 35, 23, 13]
labels = ["Family", "Phylum", "Genus", "Species"]

plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("pie.csv")
data
```

```python
plt.style.use("fivethirtyeight")
plt.pie(data['number'], labels=data['ethnicity'], wedgeprops={"edgecolor":"black", "linewidth":0.9})
plt.title("Percentage of ethnicities present at the conference")
plt.tight_layout()
plt.show()
```

#Explode

```python
plt.style.use("fivethirtyeight")

explode = [0, 0.2, 0, 0] #0.1 means 10% off the radius
plt.pie(data['number'], labels=data['ethnicity'], wedgeprops={"edgecolor":"black", "linewidth":0.9}, explode=explode, 
        shadow=True, startangle=90, autopct='%1.1f%%')

#startangle to rotate the pie around
#autopct of '%1.1f%%' will display the percentage of each segment on the pie. Without the % would be: '%1.1f'

plt.title("Percentage of ethnicities present at the conference")

plt.tight_layout()
plt.show()
```
