```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2,venn2_unweighted,venn2_circles
#could also do this: import matplotlib_venn as vplt
#then do: vplt.venn2(subsets= etc)
```

```python
items = [80, 42, 7]
labels = ["health", "Perio"]
venn2(subsets=items, set_labels=labels, set_colors=('green', 'red'), alpha=0.7)
#the above makes a venn2 weighted diagram
plt.title("2-circle venn diagram")
```

To make the circles the same size, use the unweighted

```python
venn2_unweighted(subsets=items, set_labels=labels)
```

```python
venn2_unweighted(subsets=items, set_labels=labels, alpha=0.7)
```

```python
venn2_circles(subsets=items)
```
