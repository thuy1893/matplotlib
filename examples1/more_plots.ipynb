```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize your graph (dpi specifies pixels per inch. when saving probably should use 300 if possible)
plt.figure(figsize=(5,3),dpi=100) #a dpi=300 makes the figure much bigger, try it.

#plot
#parameters for label,color,linewidth,marker,markersize,linestyle
#plt.plot(x,y,label='line1',color='red',linewidth=5,marker='.',markersize=10,linestyle='--',markeredgecolor='blue')

#alternatively you can customize using shorthand notation
#fmt = '[marker][line][color]' 
#https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.plot(x,y,'r.--',label='line1')
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.figure(figsize=(5,3),dpi=100) #a dpi=300 makes the figure much bigger, try it.
plt.plot(x,y,'r.--',label='line1')
##Line 2
# select interval we want to plot points at
x2=np.arange(0,4,0.5) #between 0 and 4, ticks every 0.5

# Plot part of the graph as line
#plt.plot(x2,x2**2,label='x^2',color='blue',linestyle='-.')
plt.plot(x2[:6],x2[:6]**2,label='x^2',color='blue')

# Plot part of the graph as dash
plt.plot(x2[5:],x2[5:]**2,label='x^2',color='blue',linestyle='--')
#title
plt.title('Our first graph!' ,fontdict={'fontname':'DejaVu Sans','fontsize':18})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X and Y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#Add a legend
plt.legend()

# Save the figure (dpi 300 is good when saving a graph at high resolution)
#plt.savefig('mygraph.png')

#show
#plt.show()
```

# Bar Chart

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)


#set patterns for the bars
bars[0].set_hatch('/')
bars[1].set_hatch('-')
bars[2].set_hatch('o')
```

```python
labels=['A','B','C']
values=[1,4,2]

bars=plt.bar(labels,values)

#same as above, but simplified
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


plt.figure(figsize=(6,4),dpi=300)
```

Load Fifa Data

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins=[40,50,60,70,80,90,100]

plt.hist(fifa.Overall ,bins=bins , color='#3fa5c4')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

Pie Charts

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']
colors=['#26c1f0','#c655f2']

plt.pie([left,right],labels=labels,colors=colors, autopct='%.2f%%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight.head()
```

```python
#noticed that at the end of the weight value is the unit 'Ibs'
#we want to remove that
fifa.Weight=[int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]
fifa.Weight.head()
```

```python
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight <200)].count()[0]
heavy=fifa[(fifa.Weight>=200)].count()[0]

weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']
explode=(.4,.2,0,0,.4)

# Plot
plt.style.use('ggplot')
plt.figure(figsize=(5,5),dpi=300)
#pctdistance to change the location of the percentage label; explode to separate the pie for better visualization
plt.pie(weights,labels=labels,autopct='%.2f %%', pctdistance=0.5, explode=explode)

plt.legend()

plt.show()
```

```python
plt.style.use('default')

barcelona= fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
liverpool=fifa.loc[fifa.Club=='Liverpool']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']

labels=['FC Barcelona','Real Madrid','Liverpool','New England Revolution']

boxes=plt.boxplot([barcelona,madrid,liverpool,revs],labels=labels, patch_artist=True , medianprops={'linewidth':2})
for box in boxes['boxes']:
    # Set edge color
    box.set(color='#1ca0ff')
    
    # Change Fill color (remember to add patch_artist=True when creating boxplot, else this won't work)
    box.set(facecolor='#1cddff')
plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
